```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

```

```python


data_frames = pd.read_csv('stock_price_dataset.csv')

```

```python
# Convert the 'Date' column to datetime
data_frames['Date'] = pd.to_datetime(data_frames['Date'], format='%y-%m-%d')

```

```python
# Extract date-related features
data_frames['Day'] = data_frames['Date'].dt.day
data_frames['Month'] = data_frames['Date'].dt.month
data_frames['Year'] = data_frames['Date'].dt.year
```

```python


# Define features and target variable
X = data_frames[['Open', 'High', 'Low', 'Volume', 'MA5', 'MA10', 'MA20', 'RSI', 'MACD', 'Signal_Line', 'SMA', 'EMA','Year','Month','Day']]
y = data_frames['Adj Close']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=300)
```

```python

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
```

```python
import pickle
pickle_out = open('predict.pkl','wb')
pickle.dump(model,pickle_out)
pickle_out.close()
```

```python
r2_score(y_test,predict)
```

```python
def modelResult(actual, predicted):
    results = []
    for a, p in zip(actual, predicted):
        results.append((a, p))
    return results

# Assuming 'actual' contains the actual values and 'predicted' contains the predicted values
results = modelResult(y_test, predict)
for actual, predicted in results:
    print("Actual:", actual, "\tPredicted:", predicted)

```
